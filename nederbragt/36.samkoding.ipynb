Heldigvis har alle programmeringsspråk smarte løsninger for vårt problem.
De løsningene bruker såkalte **løkker**.

For å komme dit at vi kan bruke en løkke må vi først justere hvordan vi
gå over alle elementene i listen vår. Vi bruker en `indeks` variabel som vi la
gå fra `0` til og med den siste indeksen i listen, her `3`.

Vi kan bruke at vi jobber i notebook på denne måten:

Først setter vi opp indeksen til `0`:

````python
indeks = 0
````

Deretter kan vi gjøre dette:

````python
print(indeks, frukter[indeks].upper())
````

````python
indeks = indeks + 1
print(indeks)
````

Nå kan vi kjøre disse to kodecellen flere ganger får å skrive ut alle

<div class="alert alert-block alert-info">
Kjør cellene flere ganger til og med `indeks` er `4` og du får en feilmelding.
</div>

Programsetningen `indeks = indeks + 1` kan virke merkelig
hvis du ser på det fra et matematisk synspunkt:
Hvordan kan `indeks` være den samme som en `indeks + 1`?
Men husk, i Python er en programsetning med `=`
et uttrykk der en variabel tilordnes en verdi.
Det som er på høyre side av `=` blir evaluert først og resultatet
variabelen til venstre for `=` settes til resultater.
Dermed kan en variabel plasseres på *begge* sider av `=` tegnet.

I dette tilfellet er variabelen `indeks` en del av beregningen på høyre side,
og dens verdi blir brukt før resultatet tilordnes `indeks` igjen.

For eksempel, hvis verdien av `indeks` er `3`,
da blir resultatet av `indeks + 1` verdien `4`
og `indeks` settes til `4` som et resultat.

## Struktur til while løkker

1. initier en løkke variable (eller flere)
2. sjekke om en betingelse er oppfylt for å kjøre løkken en gang til
3. kode som gjør det som skal repeteres
4. endre verdien av løkkevariabelen slik at etterhvert løkken avsluttes

</div>

Slik blir vår `while` løkke for å sørge for at vi får gå
over indeksene fra `0` til og med `3`:

````python
# 1. initier løkke variabelen
indeks = 0

# 2. sjekke om en betingelse er oppfylt for å kjøre løkken en gang til
while indeks < 4:
    # 3. gjør det som skal repeteres
    print(indeks)
    # 4. endre verdien av løkkevariabelen slik at etterhvert løkken avsluttes
    indeks = indeks + 1
````

<div class="alert alert-block alert-info">
Forklar løkken steg for steg
</div>

Legg merke til disse tingene:

* når du ser en på while løkke kan du tenke 'Repeat while', eller 'gjenta så lenge', her 'gjenta så lenge `indeks` er under `4`'
* linjen som starter med `while` ender med en kolon `:`. Denne betyr at det følger en eller flere linje med kode som er at det som skjer 'inne' i løkken
* koden inne i løkken er *innrykket*, noe som vi kommer til å se mye mer av i Python. Linjene som er innrykket henger sammen, og er i dette tilfelle løkkens *kodeblokk*
* i Jupyter Notebook får du innrykk 'gratis' når man ender en linje med `:`
* innrykk kan være tab eller en eller flere mellomrom, men det er mest vanlig med 4 mellomrom
* viktig at alle linjer som er inrykket har like mange mellomrom

<div class="alert alert-block alert-info">
Mulig forvirring: 
kolon `:` på slutten av `while` linje er ikke relatert til den som har med 
oppdeling av lister å gjøre: `fruits[1:3]`
</div>

Vi sier at en while-løkke består av en *betingelsestest*
etterfulgt av en *kodeblokk*, men vi kan også bruke "koden inni while-løkken"

La oss se først se litt nærmere på den delen med `indeks < 4`.
Vi kan gjøre dette:

````python
indeks = 0
indeks < 2
````

Det skrives ut `True` siden verdien av variabelen indeks faktisk er mindre enn 2.

Linjen `indeks < 2` er en *betingelse*, som evalueres til enten `True` eller
`False`.

Hva med:

````python
indeks = 3
indeks < 2
````

Nå får vi, som ventet, `False`

La oss bruke dette i vår eksempel løkke:

````python
# 1. initier løkke variabelen
indeks = 0

# 2. sjekke om en betingelse er oppfylt for å kjøre løkken en gang til
while indeks < 4:
    # 3. gjør det som skal repeteres
    print("Når indeks er", indeks, "er betingelsen:", indeks < 2)
    # 4. endre verdien av løkkevariabelen slik at etterhvert løkken avsluttes
    indeks = indeks + 1
````

Vi gjør en liten endring: nå ser vi på `indeks < 4`,
som vi også har i `while` linjen

````python
# 1. initier løkke variabelen
indeks = 0

# 2. sjekke om en betingelse er oppfylt for å kjøre løkken en gang til
while indeks < 4:
    # 3. gjør det som skal repeteres
    print("Når indeks er", indeks, "er betingelsen:", indeks < 4)
    # 4. endre verdien av løkkevariabelen slik at etterhvert løkken avsluttes
    indeks = indeks + 1
````

Her ser vi at linjen med `print` bare kjøre *så lenge* verdien av `indeks`
er mindre enn 4.

Nå skal vi bruke denne løkken til å skrive ut alle frukter
med store bokstaver:

````python
# 1. initier løkke variabelen
indeks = 0

# 2. sjekke om en betingelse er oppfylt for å kjøre løkken en gang til
while indeks < 4:
    # 3. gjør det som skal repeteres
    frukt = frukter[indeks]
    print(indeks, frukt.upper())
    # 4. endre verdien av løkkevariabelen slik at etterhvert løkken avsluttes
    indeks = indeks + 1
````

### Vanlige feil når man lager en while løkke

En av de mest vanlige feil man gjør med `while` løkken er å glemme
kolon på slutten av `while` linjen:

````python
indeks = 0
while indeks < 4
    print(indeks)
    indeks = indeks + 1
````

Dette fører til en syntax error.

Den nest vanlige feilen man gjør med `while` løkken er å glemme
*endre verdien av løkkevariabelen slik at etterhvert løkken avsluttes*.
Det man da får er en *uendelig løkke*:

```Python
        indeks = 0
        while indeks < 4:
            print(indeks)
```

Når dette skjer avsluttes løkken aldri og notebook jobber i det uendelige.
Klikk da på 'Interrupt the kernel' knappen i notebooken.
Merk at dette fører til en `KeyboardInterrupt` feilmelding
under den veldig lenge 'output'en.
Du kan dobbelklikke til venstre for det som ble skrevet ut for å gjemme det.

## Andre betingelser

Så langt har vi bare brukt `while indeks < 5:`,
dvs vi sjekker om løkke variabelen er *mindre* en ett tall.
Symbolet `<` er en av de såkalte *sammenlikningsoperatorene*.
Det er fult mulig å bruke andre betingelser også,
med bruk av de andre sammenlikningsoperatorer:

* `indeks > 5` sjekker om verdien i `indeks` er *større enn* `5`
* `indeks <= 5` sjekker om verdien i `indeks` er *mindre enn eller lik* `5`
* `indeks >= 5` sjekker om verdien i `indeks` er *større enn eller lik* `5`

Det er også mulig å sjekker om verdien til indeks *er lik* eller *er ulik*
en verdi men det har vi foreløpig ikke behov for.
De nevnes her for kompletthetens skyld:

* `indeks == 5` sjekker om verdien i `indeks` er *lik* `5`
* `indeks != 5` sjekker om verdien i `indeks` er *ulik* `5`

Merk det doble `=` likhetstegnet. Dette skyldes teste likhet mellom to ting
(`==`) fra å sette en variabelens verdi (`=`).

Vi plotter enkelt dataene våre med logaritmisk skala,
ved å legge til følgende kommando etter `plt.plot ()` kommando:


Her er vårt nye program:

````python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("ecoli_39_grader.csv")
t = list(data["t"])
E = list(data["E"])

plt.plot(t, E, "o-")
plt.yscale("log")
plt.xlabel("Time, t (minutes)")
plt.ylabel("Population size, E")
plt.title("Measured bacterial population growth")
plt.show()
````

Nå ser vi at veksten er mer eller mindre null de til rundt 50 minutter,

I tillegg blir kurven en rett linje etter det tredje punktet.
En rett linje i et plott med en logaritmisk akse tyder på
at kurven representerer eksponentiell vekst.
Dette skal v utforske i gruppetimene denne uken.

Til slutt viser vi hvordan vi kan begrense hvilken del av et plott vises.
Dette gjøres med hjelp av `plt.xlim()` og `plt.ylim()`.

Nå vi legger til `plt.xlim(60,300)` vises bare 
x aksen fra 60 minutter til 300 minutter,
som i dette tilfelle representerer den eksponentielle vekstfasen:

````python
plt.plot(t, E, "o-")
plt.yscale("log")
plt.xlim(60,300)
plt.xlabel("Time, t (minutes)")
plt.ylabel("Population size, E")
plt.title("Measured bacterial population growth")
plt.show()
````
