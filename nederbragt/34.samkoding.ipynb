Vi har nå en notebook åpen med en tom kodecelle.

```python
print("Hei verden!")
```

Vi kan kjøre koden på forskjellige måter:

1. ved å klikke på `Run` knappen øverst i notebook

2. ved å holde `shift` tasten og klikke på `enter tasten`

Hvilken du bruker er opp til deg. Men i begge tilfeller skjer det samme:

La oss nå endre koden *i den første cellen*:

```python
print("Hello world!")
```

Når vi kjører koden på nytt skrives det ut `Hello world!`.
Merk at 'telleren' ved siden av kodecellen går opp hver gang vi kjører cellen

Verdt å merke:
* koden vi skrev består av to deler

  * *funksjonen* `print()`

  * *strengen* `"Hello, world!"`

* `print` er en Python funksjon. Funksjoner finnes i mange varianter og de fleste brukes til å 'gjøre noe' for oss

* `print` funksjonen tar alt som er inne i parentesene og skriver det ut i notebooken

* `"Hei verden!"` og `"Hello, world!"` er eksempler på strenger

* tekstene er plassert mellom doble anførselstegn, dette forteller Python at vi har med en streng å gjøre

* vi kunne også ha brukte enkle anførselstegn, og det gjør ingenting så lenge samme type brukes på begge siden av en streng

* når vi skriver ut strengen bruker Python ikke anførselstegn

Hva sjer om vi glemmer et av anførselstegnene? La oss bruke den tomme cellen til det:

```python
print("Hei verden!)
```

Vi får en *feilmelding*, i dette tilfelle en `SyntaxError`.

* feilmeldingen består av flere linjer

* den første nevner en kryptisk fil, som er annerledes for hver og en av dere, og 'Line 1', dvs første linjen

* den andre kopiere linjen der feilen oppstå og indikerer der Python tror feilen er (det lille `^` symbolet)

* siste linjen er en beskjed fra Python

  * feiltypen er `SyntaxError`

  * forklaringen er `unterminated string literal (detected at line 1)`

  * dette betyr at Python kom til slutten av kodelinjen og forventet et anførselstegn, men ikke fant det

Feilmeldinger en helt vanlig og betyr ikke at noe er feil med JupyterHub,
notebooken eller deg. Det betyr at Python fant ut noe ikke stemmer.
Python sjekker koden *før* du kjører den for å finne slike feil som dette.
Men det kan også skje at Python finner feilen først når koden kjøres.

Dere kommer til å finne mange forskjellige type feil og det viktig element
av dette kurset er å forstår disse og rette dem opp.

La oss rette opp feilen og kjøre cellen på nytt:

```python
print("Hei verden!")
```

Feilmeldingen forsvinner og alt er normalt igjen.

Hva om vi glemmer det første anførselstegnet?

```python
print(Hei verden!")
```

Vi får igjen samme `SyntaxError.

Det er viktig å innimellom lagre notebooken.

# Introduksjon til variabler

Et viktig hjelpemiddel i programmeringen er variabler.
Vi bruker følgende definisjon:

> En variabel er at navn som representerer en verdi

(engelsk: "Variables are names for values")

Grunnen til at vi trenger variabler er at vi ofte må holde oversikt
over en eller flere verdier
som for eksempel

* antall ting vi skal se på

* antall ting vi har sett allerede

* løpende sum nå vi ser på flere tall

La oss bruke et eksempel.

```python
hilsen = "Hei, verden!"
```

Når vi kjører denne cellen ser vi ikke at det skjer noe,
men det gjør det allikevel:
teksten `"Hei, verden!"` representeres nå av variablen `hilsen`

Vi kan sjekke dette ved å skrive ut variabelen:

```python
print(hilsen)
```

La oss se litt nærmere på første Python kommando:

* navnet til variabelen vi vil er på venstre siden av et `=` tegn

* på høyre siden er teksten plassert mellom doble anførselstegn: `"Hei, verden!"`

* som vi har sett før forteller dette Python at vi har med en streng å gjøre

* vi sier at 'variabelen `hilsen` tilordnes (settes til) verdien `"Hei, verden!"`'

* vi sier at etter det 'variabelen `hilsen` representerer verdien `"Hei, verden!”`'

* som sagt, vi kunne også ha brukte enkle anførselstegn, og det gjør ingenting så lenge samme type brukes på begge siden av en streng

* når vi skriver ut verdien bruker Python ikke anførselstegn

* vi kan velge mange forskjell navn for variabelen, men det er noen regler:

  * navnet kan bestå av bokstaver og tall

  * første karakter kan *ikke* være et tall

  * ikke har mellomrom, men det er OK å bruke `_` tegnet

  * ikke er det samme som Python kommandoer og andre såkalte beskyttede ord; et eksempel er at du ikke bør bruke `print` som variabel navn; boken har en liste med beskyttede ord som du ikke kan bruke som navn for variabler (i ukens kapittel)

* `print` kommandoen fungere som vi har sett før, men i stedet for å skrive inn teksten som skal skrives ut direkte skriver vi inn navnet til variabelen. Python erstatter variabelen med verdien den representerer, her `"Hei, verden!"` og så skiver ut det

Merk hva som skjer når man glemmer det avsluttende anførselstegnet:

```python
hilsen = "Hei, verden!
```

Dette fører til samme `SyntaxError` som vi har sett før.
