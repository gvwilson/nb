## Oppgave 6: Volum av en celle

Man hører ofte at kroppene våre "består" av flere bakterieceller, enn våre egne celler.
Men størrelsen på disse cellene er ganske forskjellige.
Celler varierer veldig i størrelse, fra nevroner som strekker seg flere meter,
til våre relativt små røde blodlegemer på mindre enn 10 $\mu$m.

Radiusen på en menneskelig kinncelle er omtrent 50 $\mu$m,
mens en bakteriecelle ikke er større enn 5 $\mu$m.
Et vanlig estimat er en 100 ganger forskjell i radius,
det vil si at radiusen til en bakteriecelle er 0,5 mikrometer ($\mu$m),
og vi vil bruke den til denne øvelsen.

I denne øvelsen skal vi lage et Python program som skal svare på dette spørsmålet:

> Hvor mange bakterieceller ville ha samme volum som en kinncelle?

Svarene på spørsmålene nedenfor skal

* bruke variabler med fornuftige navn

* være en full setning, og skrives ut på denne måten:

```Python
        svar = # legg til svaret her
        print(svar)
```

**OBS** Ikke rund av tallene som regnes ut eller skrives ut underveis
for noen av deloppgavene!

**a)**
Skriv et Python-program som beregner volumet til en gjennomsnittlig menneskelig kinncelle,
og skriver den ut.
Volumet til en kule er gitt av formelen $\frac{4}{3}*\pi* radius^3$.

<!-- --- begin solution of exercise --- -->
**Løsningsforslag.**

````
import numpy as np
radius_kinncelle = 50
vol_kinncelle = (4/3) * np.pi * radius_kinncelle ** 3
svar = "Volumet til en kinncelle er " + str(vol_kinncelle) + " kubikk mikrometer"
print(svar)
````

**c)**
Lag nå et samlet program som beregner mengden bakterieceller som til sammen
har samme volum som en gjennomsnittlig menneskelig kinncelle.
Dette siste programmet skal altså ha alle beregningene fra de forrige deloppgavene.

<!-- --- begin solution of exercise --- -->
**Løsningsforslag.**

````
import numpy as np
radius_kinncelle = 50
vol_kinncelle = (4/3) * np.pi * radius_kinncelle ** 3
radius_bakt = 0.5
vol_bakt = (4/3) * np.pi * radius_bakt ** 3
vol_ratio = vol_kinncelle / vol_bakt
svar = "Mengden bakterieceller som har samme volum som enmenneskelig kinncelle er " + str (vol_ratio)
print(svar)
````

Hvis du avrundet tallene før du skrev ut, ville svaret være litt annerledes:

````
import numpy as np
radius_kinncelle = 50
vol_kinncelle = round((4/3) * np.pi * radius_kinncelle ** 3, 2)
radius_bakt = 0.5
vol_bakt = round((4/3) * np.pi * radius_bakt ** 3, 2)
vol_ratio = round(vol_kinncelle / vol_bakt, 2)
svar = "Mengden bakterieceller som har samme volum som enmenneskelig kinncelle er " + str (vol_ratio)
print(svar)
````
