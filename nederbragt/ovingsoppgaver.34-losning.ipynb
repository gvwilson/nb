## Oppgave 10: division

Look at the following program, and try and predict what you think the value of c is. Check your answer by running the program in Python (but guess first!)

````
a = 1
b = 4

c = a / b * 3

print(c)
````

<!-- --- begin solution of exercise --- -->
**Løsningsforslag.**
The program will print the number 0.75. The outcome, c, is a float because when we divide one integer by another integer we always get a float.

There are two operations in this program, a division and a multiplication. According to the order of operations, Python completes muliplications and divisions at the same time, with neither one taking an order of preference over the other. Thus, when there are both in one program, Python looks at the order the operations occur in the program. It therfore calculates a/ b first, and then multiplies the answer by 3. This gives us 0.75. If we had wanted Python to calculate b*3 first, we should put this operation inside parenthes.

## Oppgave 23: Explain the program

You have following program:

````
import numpy as np
cell_radius = 2
cell_volume = np.pi * cell_radius**2

print("Cell volume is:" cell_volume)
````

Without running the program, answer the following questions:

* is this program valid

* if not, explain why and correct the mistakes.

After you have answered the question,
run the prgram again to see if you were right.

<!-- --- begin answer of exercise --- -->
**Fasit.**
In line 5 (`print("Cell volume is:" cell_volume)`) there is a comma missing. Without the comma, `"Cell volume is:" cell_volume` is considered as one argument. This gives a `SyntaxError` because you have string and then a variable next to each other without any operation. Usually, dividing different things (integers, string, variable names, etc) by space only or not at all, `SyntaxError` appears.

"`"Cell volume is:"`" and "`cell_volume`" are suppose to be two arguments, and arguments in `print()` function are then just combined together. Arguments in a function in general need to be separated by comma in order to be recognized as separate arguments.

<!-- --- end answer of exercise --- -->

<!-- --- begin solution of exercise --- -->
**Løsningsforslag.**

````
import numpy as np
cell_radius = 2
cell_volume = np.pi * cell_radius**2

print("Cell volume is:", cell_volume)
````
