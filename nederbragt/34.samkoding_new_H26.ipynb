La oss nå endre koden *i den første cellen*:

```
print("Hello world!")
```

Når vi kjører koden på nytt skrives det ut `Hello world!`.
Merk at 'telleren' ved siden av kodecellen går opp hver gang vi kjører cellen

Verdt å merke:
* koden vi skrev består av to deler
  * *funksjonen* `print()`
  * *strengen* `"Hello, world!"`

* `print` er en Python funksjon. Funksjoner finnes i mange varianter og de fleste brukes til å 'gjøre noe' for oss
* `print` funksjonen tar alt som er inne i parentesene og skriver det ut i notebooken
* `"Hei verden!"` og `"Hello, world!"` er eksempler på strenger
* tekstene er plassert mellom doble anførselstegn, dette forteller Python at vi har med en streng å gjøre
* vi kunne også ha brukte enkle anførselstegn, og det gjør ingenting så lenge samme type brukes på begge siden av en streng
* når vi skriver ut strengen bruker Python ikke anførselstegn

Hva sjer om vi glemmer et av anførselstegnene? La oss bruke den tomme cellen til det:

```
print("Hei verden!)
```

Vi får en *feilmelding*, i dette tilfelle en `SyntaxError`.

* feilmeldingen består av flere linjer
* den første nevner en kryptisk fil, som er annerledes for hver og en av dere, og 'Line 1', dvs første linjen
* den andre kopiere linjen der feilen oppstå og indikerer der Python tror feilen er (det lille `^` symbolet)
* siste linjen er en beskjed fra Python
  * feiltypen er `SyntaxError`
  * forklaringen er `unterminated string literal (detected at line 1)`
  * dette betyr at Python kom til slutten av kodelinjen og forventet et anførselstegn, men ikke fant det

Feilmeldinger en helt vanlig og betyr ikke at noe er feil med JupyterHub,
notebooken eller deg. Det betyr at Python fant ut noe ikke stemmer.
Python sjekker koden *før* du kjører den for å finne slike feil som dette.
Men det kan også skje at Python finner feilen først når koden kjøres.

Dere kommer til å finne mange forskjellige type feil og det viktig element
av dette kurset er å forstår disse og rette dem opp.

La oss rette opp feilen og kjøre cellen på nytt:

```
print("Hei verden!")
```

Feilmeldingen forsvinner og alt er normalt igjen.

Hva om vi glemmer det første anførselstegnet?

```
print(Hei verden!")
```

## Biologisk case: blod-pH

I dag bygger vi ett lite program fra start til slutt.
Programmet skal bruke en målt pH-verdi fra en blodprøve,

## Steg 1: Prøven: variabler og `print()`

Vi trenger først noen grunnverdier for prøven.

```
# Case information
sample_name = "Arterial sample A"
reference_ph = 7.40
measured_ph_text = "7.10"

print("Sample:", sample_name)
print("Reference pH:", reference_ph)
print("Measured pH (text):", measured_ph_text)
```

Definisjon vi bruker i dag:

> En variabel er et navn som representerer en verdi.

Den biologiske konteksten:

* Vi lager en enkel "journal" for prøven før vi regner.
* Vi ser med en gang hva som er tekst og hva som er tall.
* måleinstrumenter kan gi svar som tallverdier, eller tall representert som strenger.

Verdt å merke:

* Til venstre for `=` står variabelnavnet, til høyre står verdien.
* Tekst må stå i anførselstegn, tall skal stå uten anførselstegn.
* Kommentarer starter med `#`, og brukes for å forklare kode til mennesker; Python ignorerer kommentarene.
* Gode variabelnavn gjør koden lettere å lese, for eksempel `reference_ph`.
* Et variabelnavn kan bestå av bokstaver, tall og liggende understrek `_ `.
* Et variabelnavn kan ikke starte med tall, og kan ikke inneholde mellomrom.
* Unngå å bruke Python-ord som `print` som variabelnavn.

## Steg 1b: Endre verdi og vanlige variabelfeiler

## Steg 2: Typekonvertering med `float()` og `type()`

For å regne med pH må vi gjøre tekst om til tall.

```
measured_ph = float(measured_ph_text)

print("Measured pH (number):", measured_ph)
print("Type of measured_ph_text:", type(measured_ph_text))
print("Type of measured_ph:", type(measured_ph))
```

Biologisk mening:

* Måleverdier kan komme som tekst fra skjema eller instrument.
* Vi må ha talltype for å bruke verdien i beregninger.

Viktig prinsipp:

* `int` betyr heltall.
* `float` betyr desimaltall.
* `str` betyr tekststreng.

Vanlig feil når vi blander typer:

```
print("Measured pH: " + measured_ph)
```

Løsning med typekonvertering:

```
print("Measured pH: " + str(measured_ph))
```

To vanlige måter å skrive ut tekst og verdier:

```
print("Measured pH:", measured_ph)
print("Measured pH: " + str(measured_ph))
```

Merk:

* Med komma settes mellomrom automatisk.
* Med `+` må vi selv styre mellomrom og bruke `str(...)` på tall.

## Steg 3: pH-avvik og avrunding

Nå regner vi forskjellen mellom målt verdi og referanse.

```
ph_difference = measured_ph - reference_ph

print("pH difference from reference:", ph_difference)
print("Rounded difference (2 decimals):", round(ph_difference, 2))
```

Legg merke til at første utskrift ofte blir noe som
`-0.3000000000000007` i stedet for `-0.3`.
Dette er ikke en feil i formelen, men en konsekvens av hvordan datamaskinen
lagrer desimaltall (`float`) i binær form.
Noen desimaltall (for eksempel `0.1`, `0.3`, `7.4`) kan ikke lagres helt presist,
så vi får små avrundingsavvik i siste desimaler.

Derfor bruker vi `round()` når vi vil ha en lesbar presentasjon av resultatet.

Biologisk mening:

* Dette er første raske vurdering av om prøven ligger over eller under referanse.

For noen Python pakker må man laste ned koden for å få tilgang til det,
mens for andre er det tilgjengelig uten videre.

Før import har vi ikke tilgang til funksjonene:

```
print(log10(100))
```

Dette gir `NameError` fordi `log10` ikke er definert ennå.

Vi importerer `numpy` og gjør om pH til $[H^+]$ med formelen
$[H^+] = 10^{-pH}$.

```
import numpy as np

h_conc_measured = 10 ** (-measured_ph)
h_conc_reference = 10 ** (-reference_ph)

print("H+ concentration (measured):", h_conc_measured)
print("H+ concentration (reference):", h_conc_reference)
```

Tall som `7.943282347242822e-08` er skrevet i *vitenskapelig notasjon*.
Her betyr `e-08` det samme som $\times 10^{-8}$.

Dermed er:

* `7.943282347242822e-08` det samme som `7.943282347242822 * 10**(-8)`
* altså et veldig lite tall: `0.00000007943282347242822`

Dette er en vanlig måte Python skriver veldig små eller veldig store flyttall på.
Notasjonen gjør resultatet mer kompakt og lettere å lese.

```
print(7.943282347242822e-08)
print(7.943282347242822 * 10 ** (-8))
```

Verdt å merke:

* `import numpy as np` betyr at vi bruker `np` som kortnavn (alias).
* Vi henter funksjoner med `np.funksjonsnavn`, for eksempel `np.log10(...)`.
* Konstantene `np.pi` og `np.e` finnes også i samme pakke.

Vi kan også sjekke at vi får tilbake pH med `np.log10()`:

```
ph_check = -np.log10(h_conc_measured)
print("pH recalculated from H+:", round(ph_check, 2))
```

Biologisk mening:

* pH-skalaen er logaritmisk.
* Små endringer i pH kan bety store endringer i $[H^+]$.

## Steg 5: Hvor mye surere enn normalt?

```
fold_acidity = h_conc_measured / h_conc_reference

message_fold = "This sample has " + str(round(fold_acidity, 2)) + " times the reference H+ concentration."
print(message_fold)
```
