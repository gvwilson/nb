#### Bessel Functions

[Bessel functions](https://en.wikipedia.org/wiki/Bessel_function) describe solutions to a particular differential equation, Bessel's equation, which describes situations such as a swinging rope, vibration in a drum, diffraction from a helix, etc.

$$
x^2 \frac{d^2 y}{dx^2} + x \frac{dy}{dx} + \left(x^2 - \alpha^2 \right)y
=
0
$$

The solutions of this equation are known as $J_n$ and $Y_n$.  The first solution, $J_n$, looks like this:

$$
J_n(x)
=
\sum_{m=0}^\infty \frac{(-1)^m}{m! \Gamma(m+n+1)} {\left(\frac{x}{2}\right)}^{2m+n}
$$
